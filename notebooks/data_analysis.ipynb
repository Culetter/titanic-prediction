```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/train.csv')
df
```

```python
df.info()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution (0 = Did not survive, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.savefig('../results/graphs/Survival_Distribution.jpg', format='jpg', dpi=300)
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survived by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived')
plt.savefig('../results/graphs/Survived_by_Tickets.jpg', format='jpg', dpi=300)
plt.show()
```

```python
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.savefig('../results/graphs/Age_Distribution.jpg', format='jpg', dpi=300)
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age by Ticket Class and Sex')
plt.xlabel('Ticket Class')
plt.ylabel('Age')
plt.savefig('../results/graphs/Age_by_Ticket_and_sex.jpg', format='jpg', dpi=300)
plt.show()
```
